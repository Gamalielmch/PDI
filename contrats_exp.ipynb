```python
import sys 
import math 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
%matplotlib inline
```

```python
#img = plt.imread(r'C:\Users\Usuario\Documents\Python\bote.jpg')
img = plt.imread(r'C:\Users\coyul\Documents\GitHub\PDI\im6.png')
fig1=plt.figure(figsize = (8,8))
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
img=img.astype('float')
inmax = np.amax(img)
c=255/(np.log(1+inmax))
Iout=np.log(img[:,:]+1)
Iout=c*Iout[:,:]

img=img.astype(np.uint8)
fig3 = plt.figure(figsize = (15,8))
a = fig3.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
a.set_title('Original Image')

Iout=Iout.astype(np.uint8)
b = fig3.add_subplot(1, 2, 2)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
b.set_title('Improvement Image')
```
