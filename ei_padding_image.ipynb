```python
n=np.array([15,15])
sz=im.shape
imt=im
temp=np.zeros((n[0],sz[1]))
temp[0:n[0],:]=imt[sz[0]-n[0]:sz[0],:]
print(temp)
print(np.flip(temp,1))
```

```python

temp=np.zeros((1,sz[1]))
temp[0,:]=imt[sz[0]-1,:] 
temp=np.repeat(temp, n[0], axis=0)
imt=np.vstack((imt,temp))
temp=np.zeros((1,sz[1]))
temp[0,:]=imt[0,:]
temp=np.repeat(temp, n[0], axis=0)
imt=np.vstack((temp,imt))

sz=imt.shape
temp=np.zeros((sz[0],1))
temp[:,0]=imt[:,sz[1]-1] 
temp=np.repeat(temp, n[1], axis=1)
imt=np.hstack((imt,temp))  

temp=np.zeros((sz[0],1))
temp[:,0]=imt[:,0]
temp=np.repeat(temp, n[1], axis=1)
imt=np.hstack((temp,imt))


fig2=plt.figure(figsize = (6,6))
plt.imshow(imt, cmap='gray',vmin=0,vmax=255);
plt.axis('off');
```
