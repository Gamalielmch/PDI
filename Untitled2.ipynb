```python
t=os.listdir("C:/Users/Gama/Documents/Github/PDI/TalentJ1/")
for i in range(480,490):
    plt.figure(figsize = (3,3))
    path= "C:/Users/Gama/Documents/Github/PDI/TalentJ1/"  +str(i)+".jpg"
    I=Image.open(path)
    I=np.array(I)
    I=I[:,:,1]
    I=I.astype(int)
    mini=np.amin(I)
    maxi=np.amax(I)
    pi, bins,=np.histogram(I.ravel(), np.arange(256))
    c_min=80
    pi[0:c_min]=0
    pi=pi/np.sum(pi)
    bins = np.linspace(0,255,255)
    sgm2=[]
    for t in range(mini,maxi):
        w0= np.sum(pi[mini:t])+0.0000001
        w1= 1-w0
        miu0=np.sum(np.arange(t)*pi[0:t])/w0
        miu1=np.sum(np.arange(t+1,maxi)*pi[t+1:maxi])/w1
        sgm2=np.append(sgm2,w0*w1*(miu0-miu1)**2)
    umbral=np.argmax(sgm2)+c_min
    si=I.shape
    S=np.zeros(si)
    for x in range(si[0]):
        for y in range(si[1]):
            if I[x,y]>=umbral:
                S[x,y]=255
    S=S.astype(int)
    
    plt.imshow(I, cmap='gray',vmin=0,vmax=255);
    plt.axis('off');
    plt.show
    
```
