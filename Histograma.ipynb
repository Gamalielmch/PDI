```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r'C:\Users\coyul\Documents\GitHub\PDI\imcolor4.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
fig1 = plt.figure(figsize = (6,6))
a = fig1.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title('Image');

b = fig1.add_subplot(1, 2, 2)
plt.hist(img.ravel(),256,[0,256]); 
plt.show()
```

```python
inmax = np.amax(img)
inmin = np.amin(img)
print(inmin,inmax)
m=255/(inmax-inmin)
b=-m*inmin
Iout=img*m+b
print(np.amin(Iout),np.amax(Iout))
```

```python
fig1 = plt.figure(figsize = (6,6))
a = fig1.add_subplot(1, 2, 1)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title('Image');

b = fig1.add_subplot(1, 2, 2)
plt.hist(Iout.ravel(),256,[0,256]); 
plt.show()
```

```python
fig1 = plt.figure(figsize = (6,6))
a = fig1.add_subplot(1, 2, 1)
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title('Original image');

b = fig1.add_subplot(1, 2, 2)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
b.set_title('Enhanced Image');
```
