```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import padding_image as pim
from rgb2gray import rgbtogray as r2g
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
```

```python
############# LOADER AND GRAYSCALE CONVERTER IMAGE
root = tk.Tk()
root.withdraw()
file_path = filedialog.askopenfilename()
if file_path=='':
    messagebox.showinfo("Warning", "Image not loaded")
else:
    im=Image.open(file_path)
    im=np.array(im)
    im=im.astype(int)
    if im.ndim>2:
        im=r2g(im)

    fig1 = plt.figure(figsize = (6,6))
    a = fig1.add_subplot(1, 1, 1)
    plt.imshow(im, cmap='gray',vmin=0,vmax=255)
    plt.axis('off');
    a.set_title('Original Image');
```

```python
################ FILTERING
imt=im
w=np.array([[1,2,3,4],[5,6,7,8],[11,12,13,14]])
n=w.shape
sz=imt.shape
n=np.asarray(n)
n_p=np.zeros((2))
if n[0]%2 > 0:
    n_p[0]=(n[0]-1)/2
else:
    n_p[0]=n[0]/2
if n[1]%2 > 0:
    n_p[1]=(n[1]-1)/2
else:
    n_p[1]=n[1]/2

n_p=n_p.astype(int)
im_p= pim.padding(imt,n_p,"zeros")
fig2 = plt.figure(figsize = (16,16))
b = fig2.add_subplot(1, 1, 1)
plt.imshow(im_p, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
b.set_title('Padding Image');


```
