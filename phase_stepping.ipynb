```python
#phase stepping
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from rgb2gray import rgbtogray
import os
%matplotlib inline
```

```python
I=np.zeros((512,512))
I=I.astype(int)
N=3
print(np.pi)
x=np.arange(0,512)
n=0
s=np.cos(((2*np.pi*x)/300)+2*np.pi*n/N) 
plt.plot(s) 
plt.show
```

```python
for n in range(3):
    s=128+127*np.cos(((2*np.pi*x)/80)+2*np.pi*2/N) 
    for i in range(512):
        for j in range(512):
            I[i, j]= s[j]        
fig1=plt.figure(figsize = (10,10))
plt.imshow(I, cmap='gray',vmin=0,vmax=255);
plt.axis('off');
im = Image.fromarray(np.uint8(I))
im1 = im.save("step3.jpg") 
```
