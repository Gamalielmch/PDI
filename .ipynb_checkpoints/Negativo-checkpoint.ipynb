```python
import sys 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
%matplotlib inline
```

```python
#img = plt.imread(r'C:\Users\Usuario\Documents\Python\bote.jpg')
img = plt.imread(r'C:\Users\coyul\Documents\GitHub\PDI\im3.jpeg')
fig1=plt.figure(figsize = (8,8))
plt.imshow(img, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```

```python
inmax = np.amax(img)
inmin = np.amin(img)

Lout=0;
Hout=255;
m=(Hout-Lout)/(inmax-inmin);
b=Lout-m*inmin;
Iout=img*m+b;

fig3 = plt.figure(figsize = (15,8))
a = fig3.add_subplot(1, 3, 1)
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
a.set_title('Improvement Image')

Iout=Iout.astype('float');
Iout=255-Iout;
b = fig3.add_subplot(1, 3, 2)
b.set_title('Negative Image of Improvement')
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
Iout=255-img.astype('float');
c = fig3.add_subplot(1, 3, 3)
c.set_title('Negative Image of Original')
plt.imshow(Iout, cmap='gray',vmin=0,vmax=255)
plt.axis('off');
```
