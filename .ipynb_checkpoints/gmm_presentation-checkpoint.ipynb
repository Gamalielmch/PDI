# Mezclas de Modelos Gaussianos  (Guassian Mixture Models)
<div style="font-size: 20px";  >
Los modelos de mezcla gaussiana son un modelo probabilístico para representar subpoblaciones normalmente distribuidas dentro de una población general. Los modelos de mezclas, en general, no requieren saber a qué subpoblación pertenece un punto de datos, lo que permite que el modelo aprenda las subpoblaciones automáticamente. Como la asignación de subpoblaciones no se conoce, esto constituye una forma de aprendizaje no supervisado.
    
Un modelo de mezcla gaussiana se parametriza mediante dos tipos de valores, los pesos de los componentes de la mezcla y las medias y varianzas / covarianzas del componente.
</div > 

## Función de distribución Gaussiana univariable

<div style="font-size: 25px", align="center">
$ N (x \vert \mu_{i}, \sigma_{i}) = \frac{1}{\sigma_{i} \sqrt{2 \pi}} exp ^{\left( - \frac{  { (x-\mu_{i}) }^2} { 2 \sigma_{i}^{2} }    \right)} $
</div > 

```python
import numpy as np
import time 
import matplotlib.pyplot as plt
from PIL import Image
```

```python
L=np.arange(256)
m=128
sgm=5
g=np.zeros((256))
print((1 / (sgm *np.sqrt(2*np.pi) )) )
for x in range(256):
    g[x]= (1 / (sgm *np.sqrt(2*np.pi) )) * np.exp( -(x-m)**2 / (2*sgm**2) )
    
#print(g)

fig1=plt.figure(figsize = (6,6))
plt.plot(L,g);
 
```

## Función de distribución Gaussiana univariable

<div style="font-size: 25px", align="center">
$ N (x \vert \mu_{i}, \sigma_{i}) = \frac{1}{\sigma_{i} \sqrt{2 \pi}} exp ^{\left( - \frac{  { (x-\mu_{i}) }^2} { 2 \sigma_{i}^{2} }    \right)} $
</div > 

## Mezcla de modelos Gaussianos 
<div style="font-size: 25px", align="center">
$p(x) = \displaystyle \sum_{i=1}^{K} \phi_{i} N (x \vert \mu_{i}, \sigma_{i}) $
</div >
<div style="font-size: 20px; margin-top: 1em;">
donde $k=1,2,\dots ,K $ son los modelos y $ \sum_{i=1}^{K} \phi_{i} =1$
</div >   


<div style="font-size: 20px; margin-top: 1em;" >
Para encontrar las medias y varianzas que mejor se ajusta a la distribución se realiza mediante un método iterativo llamado Expectation maximization (EM). Esta es una técnica numérica para la estimación de máxima verosimilitud. El algorimto EM tiene la propiedad de que la verosimilitud aumenta estrictamente con cada iteración, lo que significa que se garantiza que se acerque a un máximo local o global.
</div >

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = np.round (0.2989 * r + 0.5870 * g + 0.1140 * b)

    return gray


im2=Image.open('C:/Users/Usuario/Documents/MATLAB/PDI/subject05_80.jpg')
im2=np.array(im2)
im2=im2.astype(int)
fig1=plt.figure(figsize = (6,6))
plt.imshow(im2, cmap='gray',vmin=0,vmax=255);
plt.axis('off');
```

### Definiendo número de clases, medias y varianzas iniciales 

```python
k=5
m=np.linspace(5,250,k)
sgm=np.ones(k)*5
phik=vari=np.ones(k)/k
print ("medias iniciales:    ",m )
print ("varianzas iniciales:  ",sgm )
print ("Pesos iniciales:  ",phik )
fig=plt.figure(figsize = (6,6))
L=np.arange(256)
g=np.zeros((256,k))
for y in range(3):   
    for x in range(256):
        for ki in range(k):
             g[x,ki]= (1 / (sgm[ki] *np.sqrt(2*np.pi) )) * np.exp( -(x-m[ki])**2 / (2*sgm[ki]**2) )
    plt.plot(L,g);
    plt.plot(L,fx, color='xkcd:black');
  
```

### Algoritmo EM

#### Expectation (E) Step:
<div style="font-size: 20px; margin-top: 1em;" >
Primero se calcula la probabilidad $\gamma_{i k}$ de que el pixel $x_{i}$ haya sido generado por la clase  $C_{k}$, esto es $\gamma_{ik}=p(C_{k} \vert x_{i}, \phi, \mu, \sigma )$
</div > 


<div style="font-size: 25px; margin-top: 2em;", align="center">
$\gamma_{ik} =  \frac{\phi_{k}   N(x_{i}\vert \mu_{k}, \sigma_{k} }  { \sum_{j=1}^{K} \phi_{j} N(x_{i}\vert \mu_{j}, \sigma_{j})  }$
</div >
    
#### Maximization (M) Step:   
<div style="font-size: 20px; margin-top: 1em;" >
Empleando  $\gamma_{i k}$ se calcula lo siguiente para cada modelo k
</div >

<div style="font-size: 25px; margin-top: 2em;", align="center">
$\phi_{k} = \sum_{i=1}^{N} \frac{\gamma_{ik}}{N}$
</div >

<div style="font-size: 25px; margin-top: 2em;", align="center">
$\mu_{k} =  \frac{\sum_{i=1}^{N} \gamma_{ik} x_{i}}{\sum_{i=1}^{N}{\gamma_{ik}}}$    
</div >
    
<div style="font-size: 25px; margin-top: 2em;", align="center">
$\sigma_{k}^{2} =  \frac{\sum_{i=1}^{N} \gamma_{ik} (x_{i}-\mu_{k})^{2}}{\sum_{i=1}^{N}{\gamma_{ik}}}$       
</div >

```python
k=5
g=np.zeros((256,k))
mini=np.amin(im2)
maxi=np.amax(im2)
ra=maxi-mini
m=np.linspace(mini+0.25*ra,maxi-0.25*ra,k)
# print(m)
sgm=np.ones(k)*5
phik=vari=np.ones(k)/k
im3=im2.flatten()
N=len(im3);
gamma=np.zeros((len(im3),k))
# print(gamma.shape)
# print(im3.shape)

print('Start')
#for ite in range(2):    
for ite in range(80): 
    for x in range(len(im3)):
        suma=0;
        for ki in range(k):
            suma=suma+ (phik[ki] / (sgm[ki] *np.sqrt(2*np.pi))) * np.exp( -(im3[x]-m[ki])**2 / (2*sgm[ki]**2) ) 
    
        for ki in range(k):
            gamma[x,ki]=((phik[ki] /(sgm[ki] *np.sqrt(2*np.pi)) * np.exp( -(im3[x]-m[ki])**2 / (2*sgm[ki]**2) ) )/suma)
            
            
            
    for ki in range(k):
        suma=np.sum(gamma[:,ki])
        phik[ki] = suma /N
        m[ki]= np.dot(gamma[:,ki],im3[:])/ suma
        sgm[ki]=np.sqrt  ( np.dot(gamma[:,ki], (im3[:]-m[ki])**2) /suma )
#    print(m)
#     fig=plt.figure(figsize = (6,6))
#     for xx in range(256):
#          for ki in range(k):
#             g[xx,ki]= phik[ki]*(1 / (sgm[ki] *np.sqrt(2*np.pi) )) * np.exp( -(xx-m[ki])**2 / (2*sgm[ki]**2) )
#     plt.plot(L,g); 
#    # plt.plot(L,g.sum(axis=1),color='xkcd:red'); 
#     plt.plot(L,fx, color='xkcd:black');  

print('FINISH')
print(phik)

fig=plt.figure(figsize = (15,15))
for xx in range(256):
    for ki in range(k):
        g[xx,ki]= phik[ki]*(1 / (sgm[ki] *np.sqrt(2*np.pi) )) * np.exp( -(xx-m[ki])**2 / (2*sgm[ki]**2) )
plt.plot(L,g); 
plt.plot(L,g.sum(axis=1),color='xkcd:red'); 
plt.plot(L,fx, color='xkcd:black');  
```
