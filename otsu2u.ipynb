```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = np.round (0.2989 * r + 0.5870 * g + 0.1140 * b)

    return gray


#im2=Image.open('C:/Users/Usuario/Documents/MATLAB/PDI/subject05_80.jpg')
im2=Image.open('C:/Users/Usuario/Documents/MATLAB/PDI/im_contrast_2.tiff')
im2=np.array(im2)
#im2=rgb2gray(im2)
im2=im2.astype(int)
fig1=plt.figure(figsize = (6,6))
#print(im2.shape)
plt.imshow(im2, cmap='gray',vmin=0,vmax=255);
plt.axis('off');
```

```python
i=np.arange(256)
mini=np.amin(im2)
print(mini)
maxi=np.amax(im2)
print(maxi)
vari=np.zeros(256)
#mc1=np.zeros(255)
#mc2=np.zeros(255)
#b = im2.ravel();
p, bins, bars=plt.hist(im2.ravel(), bins=256, range=(0.0, 256));
p =p/sum(p)
```

```python
maximo=0;
MT=np.inner(i,p)
for j in range(mini, maxi-2):
    for k in range(j+1,maxi-1):
        k1=np.sum(p[mini:j])+0.1e-13
        mc1=np.inner(i[mini:j],p[mini:j])/k1;
        k2=np.sum(p[j+1:k])+0.1e-13
        mc2=np.inner(i[j+1:k],p[j+1:k])/k2;
        k3=np.sum(p[k+1:maxi])+0.1e-13
        mc3=np.inner(i[k+1:maxi],p[k+1:maxi])/k3;
        vari=k1*(mc1-MT)**2 + k2*(mc2-MT)**2 + k3*(mc3-MT)**2;
        if vari>maximo:
            u=j
            v=k
            maximo=vari
print(u,v)  


fig = plt.figure(figsize = (14,14))

a = fig.add_subplot(1, 4, 1)
plt.imshow(im2, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
a.set_title('Original Image')


b = fig.add_subplot(1, 4, 2)
plt.imshow((im2<=u).astype(int)*255, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
b.set_title('C1 image')

c = fig.add_subplot(1, 4, 3)
plt.imshow((np.logical_and(im2>u,im2<=v)) .astype(int)*255, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
b.set_title('C2 image')

c = fig.add_subplot(1, 4, 4)
plt.imshow((im2>v).astype(int)*255, cmap='gray',vmin=0,vmax=255)
plt.axis('off')
b.set_title('C4 image')
```
